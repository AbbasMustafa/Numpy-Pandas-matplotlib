```python
# Day 1

import numpy as np
```

```python
# Create numpy array using python list

x = np.array([1,2,3])
print( "Dimension: {}".format(x.shape) )
print( "data Type: {}".format(x.dtype) )
print( "type: {}".format(type(x)) )
print( "Values: {}".format(x) )
```

```python
# numpy array hold string

x = np.array(["hello","World"])
print( "Dimension: {}".format(x.shape) )
print( "data Type: {}".format(x.dtype) )
print( "type: {}".format(type(x)) )
print( "Values: {}".format(x) )
```

```python
# numpy can't hold different datatype it'll interpt all elements as string

x = np.array([1,2,"World"])
print( "Dimension: {}".format(x.shape) )
print( "data Type: {}".format(x.dtype) )
print( "type: {}".format(type(x)) )
print( "Values: {}".format(x) )
```

```python
# numpy can't hold different datatype it'll interpt all elements as float (Upcasting)

x = np.array([1,2.0,3,4])
print( "Dimension: {}".format(x.shape) )
print( "data Type: {}".format(x.dtype) )
print( "type: {}".format(type(x)) )
print( "Values: {}".format(x) )
```

```python
# we can set dtype druing initilization of ndarray by using keyword dtype

x = np.array([1,2,3,4],dtype=np.float64)
print( "Dimension: {}".format(x.shape) )
print( "data Type: {}".format(x.dtype) )
print( "type: {}".format(type(x)) )
print( "Values: {}".format(x) )
```

```python
# multi dimension array in numpy

x = np.array([[1,2],[1,2],[1,2]])
print( "Dimension: {}".format(x.shape) )
print( "data Type: {}".format(x.dtype) )
print( "type: {}".format(type(x)) )
print( "Values: \n{}".format(x) )
```

```python
# we can save numpy array for future use un different program as well
# it's save in .npy extension file

x = np.array(["This","data","is","saved"])
np.save('my_array',x)
```

```python
# Load array from 'my_array.npy' file
# note we must put .npy on the last of the file name

y = np.load('my_array.npy')
print( "Dimension: {}".format(y.shape) )
print( "data Type: {}".format(y.dtype) )
print( "type: {}".format(type(y)) )
print( "Values: \n{}".format(y) )
```

```python
# numpy array using built-in function
# Creating ndarray full of zeros

x = np.zeros((3,3))
print( "Dimension: {}".format(x.shape) )
print( "data Type: {}".format(x.dtype) )
print( "type: {}".format(type(x)) )
print( "Values: \n{}".format(x) )
```

```python
# ndarray full of ones 

x = np.ones((3,4),dtype=np.int64)
print( "Dimension: {}".format(x.shape) )
print( "data Type: {}".format(x.dtype) )
print( "type: {}".format(type(x)) )
print( "Values: \n{}".format(x) )
```

```python
# ndarray full of any number you want 

x = np.full((3,4),5)
print( "Dimension: {}".format(x.shape) )
print( "data Type: {}".format(x.dtype) )
print( "type: {}".format(type(x)) )
print( "Values: \n{}".format(x) )
```

```python
# creating identity matrix

x = np.eye(3,dtype=np.int64)
print( "Dimension: {}".format(x.shape) )
print( "data Type: {}".format(x.dtype) )
print( "type: {}".format(type(x)) )
print( "Values: \n{}".format(x) )
```

```python
# diagnal value with our own choice

x = np.diag([1,2,3,4])
print( "Dimension: {}".format(x.shape) )
print( "data Type: {}".format(x.dtype) )
print( "type: {}".format(type(x)) )
print( "Values: \n{}".format(x) )
```

```python
# evenly space values with a given interval

x = np.arange(3)
print( "Dimension: {}".format(x.shape) )
print( "data Type: {}".format(x.dtype) )
print( "type: {}".format(type(x)) )
print( "Values: \n{}".format(x) )

print()
 
x = np.arange(3,10)
print( "Dimension: {}".format(x.shape) )
print( "data Type: {}".format(x.dtype) )
print( "type: {}".format(type(x)) )
print( "Values: \n{}".format(x) )

print()
 
x = np.arange(3,10,2)
print( "Dimension: {}".format(x.shape) )
print( "data Type: {}".format(x.dtype) )
print( "type: {}".format(type(x)) )
print( "Values: \n{}".format(x) )
```

```python
# np.linspace function for interval between fixed values start and stop
# Default generate 50 values between given values [start and Stop]

x = np.linspace(0,10)
print( "Dimension: {}".format(x.shape) )
print( "data Type: {}".format(x.dtype) )
print( "type: {}".format(type(x)) )
print( "Values: \n{}".format(x) )
```

```python
# we can set how many value we want between interval [start,stop]

x = np.linspace(0,10,6)
print( "Dimension: {}".format(x.shape) )
print( "data Type: {}".format(x.dtype) )
print( "type: {}".format(type(x)) )
print( "Values: \n{}".format(x) )
```

```python
# we can set that stop value will be exclusive by suing keyword endpoint=False

x = np.linspace(0,10,5,endpoint=False)
print( "Dimension: {}".format(x.shape) )
print( "data Type: {}".format(x.dtype) )
print( "type: {}".format(type(x)) )
print( "Values: \n{}".format(x) )
```

```python
# we can create 1darray into ndarray by using np.reshape

x = np.arange(0,10)
x = np.reshape(x,(2,5))
print( "Dimension: {}".format(x.shape) )
print( "data Type: {}".format(x.dtype) )
print( "type: {}".format(type(x)) )
print( "Values: \n{}".format(x) )
```

```python
# we can do this in a signle line by this

x = np.arange(0,10).reshape(2,5)
print( "Dimension: {}".format(x.shape) )
print( "data Type: {}".format(x.dtype) )
print( "type: {}".format(type(x)) )
print( "Values: \n{}".format(x) )
```

```python
# generate random number using numpy values between [0.0,1.0)

x = np.random.random((2,2))
print( "Dimension: {}".format(x.shape) )
print( "data Type: {}".format(x.dtype) )
print( "type: {}".format(type(x)) )
print( "Values: \n{}".format(x) )
```

```python
# generate random number using numpy values between user wants any number [start,stop)

x = np.random.randint(1,3,(4,4))
print( "Dimension: {}".format(x.shape) )
print( "data Type: {}".format(x.dtype) )
print( "type: {}".format(type(x)) )
print( "Values: \n{}".format(x) )
```

```python
# generate random number that satisfy the certain statistical properties

x = np.random.normal(0,0.1,size=(100,100))
print( "mean: {}".format( x.mean()) )
print( "max: {}".format( x.max()) )
print( "min: {}".format( x.min()) )
print( "(x<0): {}".format( (x<0).sum()) )
print( "(x>0): {}".format( (x>0).sum()) )
```
