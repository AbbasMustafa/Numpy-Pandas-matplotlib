```python
import numpy as np
```

```python
# We can do basic mathematical oprtion using numpy of 1darray

x = np.arange(10)
y = np.arange(10,20)

print("x + y: \n{}".format(x+y))
print()
print("np.add(x,y): \n{}".format(np.add(x,y)))
print()
print("x - y: \n{}".format(x-y))
print()
print("np.subtract(x,y): \n{}".format(np.subtract(x,y)))
print()
print("x * y: \n{}".format(x*y))
print()
print("np.multiply(x,y): \n{}".format(np.multiply(x,y)))
print()
print("x / y: \n{}".format(x/y))
print()
print("np.divide(x,y): \n{}".format(np.divide(x,y)))
print()
```

```python
# We can do basic mathematical oprtion using numpy of ndarray

x = np.arange(10).reshape(2,5)
y = np.arange(10,20).reshape(2,5)

print("x + y: \n{}".format(x+y))
print()
print("np.add(x,y): \n{}".format(np.add(x,y)))
print()
print("x - y: \n{}".format(x-y))
print()
print("np.subtract(x,y): \n{}".format(np.subtract(x,y)))
print()
print("x * y: \n{}".format(x*y))
print()
print("np.multiply(x,y): \n{}".format(np.multiply(x,y)))
print()
print("x / y: \n{}".format(x/y))
print()
print("np.divide(x,y): \n{}".format(np.divide(x,y)))
print()
```

```python
# Mathematical Function provided by numpy

x = np.array([[1,2],[3,4]])

print("exp: \n{}".format(np.exp(x)))
print()
print("sqrt: \n{}".format(np.sqrt(x)))
print()
print("square: \n{}".format(np.power(x,2)))
```

```python
# statisticall Function provided by numpy

# Mean
x = np.array([[1,2],[3,4]])

print("x: \n{}".format(x))
print()
print("Mean of array: \n{}".format(x.mean()))
print()
print("Mean of columns: \n{}".format(x.mean(axis=0)))
print()
print("Mean of rows: \n{}".format(x.mean(axis=1)))
print()
```

```python
# statisticall Function provided by numpy

# sum
x = np.array([[1,2],[3,4]])

print("x: \n{}".format(x))
print()
print("sum of array: \n{}".format(x.sum()))
print()
print("sum of columns: \n{}".format(x.sum(axis=0)))
print()
print("sum of rows: \n{}".format(x.sum(axis=1)))
print()
```

```python
# statisticall Function provided by numpy

# std
x = np.array([[1,2],[3,4]])

print("x: \n{}".format(x))
print()
print("std of array: \n{}".format(x.std()))
print()
print("std of columns: \n{}".format(x.std(axis=0)))
print()
print("std of rows: \n{}".format(x.std(axis=1)))
print()
```

```python
# statisticall Function provided by numpy

# median must use np.median()
x = np.array([[1,2],[3,4]])

print("x: \n{}".format(x))
print()
print("median of array: \n{}".format(np.median(x)))
print()
print("median of columns: \n{}".format(np.median(x,axis=0)))
print()
print("median of rows: \n{}".format(np.median(x,axis=1)))
print()
```

```python
# statisticall Function provided by numpy

# max 
x = np.array([[1,2],[3,4]])

print("x: \n{}".format(x))
print()
print("max of array: \n{}".format(x.max()))
print()
print("max of columns: \n{}".format(x.max(axis=0)))
print()
print("max of rows: \n{}".format(x.max(axis=1)))
print()

#min
print("min of array: \n{}".format(x.min()))
print()
print("min of columns: \n{}".format(x.min(axis=0)))
print()
print("min of rows: \n{}".format(x.min(axis=1)))
print()
```

```python
# BroadCast Scaler to Matrix

x = np.array([[1,2],[3,4]])

print("x: \n{}".format(x))
print()
print("3 + x: \n{}".format(np.add(3,x)))
print()
print("3 - x: \n{}".format(np.subtract(3,x)))
print()
print("3 * x: \n{}".format(3*x))
print()
print("3 / x: \n{}".format(3/x))
print()
```

```python
# BroadCast Matrix to Matrix

x = np.array([1,2,3])

y = np.arange(9).reshape(3,3)


print("x:\n{}".format(x))
print()
print("y:\n{}".format(y))
print()
print("x add into y each row:\n{}".format(x+y))
print()
print("x add into y each columns:\n{}".format(x.reshape(-1,1) + y))
print()
```
